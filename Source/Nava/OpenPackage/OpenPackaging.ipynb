## Open a Xlsx file

```polyglot-notebook
open System
open System.IO

let dataPath = Path.Combine(__SOURCE_DIRECTORY__,"..","..","..","Data") |> Path.GetFullPath |> DirectoryInfo
let testXlsxFile = dataPath.GetFiles("Test.xlsx") |> Seq.head
open System.Xml
open Nava.Excel.Common
open FsToolkit.ErrorHandling
```

```polyglot-notebook
let stream = testXlsxFile.OpenRead()
//workbook.Sheets |> Seq.iter(fun keyValue -> keyValue.K |> printf "%s")//(fun keyValue -> workbook.Relations[keyValue.Value].ToString() |> printf "%s")
match Workbook.openWithOptions OpenPackage.defaultOpenOptions stream with
| Ok workbook ->
    workbook.Display() |> ignore
    (workbook.Worksheets["Sheet1"] |> Worksheet.enumerateRows |> Seq.take 20).DisplayTable()
    workbook.Dispose()
| _ -> ()
```
